### Linear regression on simple data sets

- Write a function that given the input data returns the fitting parameters $a_1$ and $a_2$. The function template is:

```python

```

- Test your function on a simple set of data: plot the data and the fitted function

````python
def lin_reg(x_data, y_data):
    """If you give this this function x_data and the corresponding y_data,
    it'll return you a0 and a1 for linear regression"""
    
    x_data_avg  = np.mean(x_data)
    y_data_avg  = np.mean(y_data)
    
    N = np.squeeze(x_data.shape)
    
    
    s1 = np.sum(x_data*y_data) - N*x_data_avg*y_data_avg
    
    s2 = np.sum(x_data**2) - N*x_data_avg*x_data_avg

    a1 = s1/s2

    a0 = y_data_avg - a1*x_data_avg

    return (a0, a1)

# Testing the function

#Generating Data
x = linspace(1,100,1000)
y = x**2

a0test, a1test = lin_reg(x,y)
lin_fit = a1test*x + a0test


#Plotting the Data
figure(figsize=(20,7))
plot(x, y, 'r',label=r"$ y = x^2 $")
plot(x, lin_fit, 'b',label=r" Linear regression fit of the Data ")
legend(loc = 2)
xlabel('x')
ylabel('y')
title("x   vs   function 1   &   function 2")
show()
````

#### So, we see that the function lin_reg(x,y) is working perfectly fine!
